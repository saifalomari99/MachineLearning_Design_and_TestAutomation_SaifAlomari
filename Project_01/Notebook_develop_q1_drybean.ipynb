````python
import openai 

openai.api_key = "your_key_here"


# ----- define a helper function that sends a prompt to GPT
def call_llm(prompt: str) -> str:
    response = openai.chat.completions.create(
        model="gpt-4o-mini",
        messages=[
            {"role": "system", "content": "You are a Python coding assistant."},
            {"role": "user", "content": prompt},
        ]
    )
    return response.choices[0].message.content
````

````python
prompt = """
You are an expert at writing a Python code.
Follow the instructions below to generate a Python code.

1. Use the pandas library.
2. Load the csv file '/autograder/source/drybean.csv' into a pandas DataFrame named df.
3. Store the first 5 rows of df in a variable named answer_1.
4. Store a list of the column names of df in a variable named answer_2.
5. Store the shape of df in a variable named answer_3.
6. Store the number of unique values of each column in a variable named answer_4.

Do not include any explanations or comments, only return the Python code.
Do not include ```python.
"""
````

````python
# === 2) Save prompt to the required txt file ===
PROMPT_FILENAME = "hw1_q1_prompt.txt"
with open(PROMPT_FILENAME, "w", encoding="utf-8") as f:
    f.write(prompt)
print(f"✅ Saved prompt to {PROMPT_FILENAME}")
````

````python
# ---- generate the code
generated_code = call_llm(prompt)
print(generated_code)
````
